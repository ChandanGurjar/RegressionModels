let's check for the multicollinearity we need to take the help of statsmodels library 
we'll use vif function variance inflation factor 



```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables=raw_data[['Age','BMI']]
vif=pd.DataFrame()
vif['Vif']=[variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif['Features']=variables.columns
vif
```

```python
log_Insurance_Charges=np.log(data_with_dummies['Insurance Charges'])
data_with_dummies['log_Insurance_Charges']=log_Insurance_Charges
data_with_dummies
```

```python
data_with_dummies.describe(include='all')
```

Linear Regression Model 


```python
data_with_dummies=data_with_dummies.drop(['Insurance Charges'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
target=data_with_dummies['log_Insurance_Charges']
inputs=data_with_dummies.drop(['log_Insurance_Charges'],axis=1)
```

```python
scaler.fit(inputs)
```

```python
inputs_scaled=scaler.transform(inputs)
```

```python
inputs_scaled.shape
```

```python
target.shape
```

```python
reg=LinearRegression()
```

```python
# now divide the data for train and test by using the train_test_split

x_train,x_test,y_train,y_test=train_test_split(inputs_scaled,target,test_size=0.2,random_state=365)
```

```python
reg.fit(x_train,y_train)
```

```python
y_hat=reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('prediction(y_hat)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# residual plot
sns.distplot(y_train-y_hat)
```

 it show that there are lot of times the predictions are much lower than the targets

```python
# r-score 

reg.score(x_train,y_train)
```

not bad our model is predicting 75% variablity which is not that bad 

```python
reg_summary=pd.DataFrame(inputs.columns.values,columns=['Features'])
reg_summary['Weights']=reg.coef_
reg_summary
```

so the weights shows that the most important feature is the smoker who are yes  then comes the age it seems like the region doesn't have that much predictive power  and the sex_male is negative which shows that female have high insurance charges than male 

```python
# testing 

y_hat_test=reg.predict(x_test)

```

```python
plt.scatter(y_test,y_hat_test,alpha=0.2,c='orange',cmap='green')
plt.xlabel('Target(y_test)',size=18)
plt.ylabel('Predictions(y_hat_test',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf=pd.DataFrame(np.exp(y_hat_test),columns=['Prediction'])
df_pf.head()
```

```python
df_pf['Target']=np.exp(y_test.reset_index(drop=True))
df_pf.head()
```

```python
# residual 
df_pf['Residual']=df_pf['Target']-df_pf['Prediction']
```

```python
# residual in % 
df_pf['Difference%']=np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```
