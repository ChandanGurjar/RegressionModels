```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
sns.set()

data=pd.read_csv('Startups.csv')
data.head(10)
```

```python
y=data['Profit']
x=data['Marketing Expenditure']

x1=sm.add_constant(x)
result1=sm.OLS(y,x1).fit()
result1.summary()
```

```python
my_data=pd.DataFrame({'Const':[1],'Marketing Expenditure':[325000]})

my_data=my_data[['Const','Marketing Expenditure']]
my_data


```

```python
result1.predict(my_data)
```

```python
y1=data['Profit']
x2=data[['R&D Expenditure','Administration Expenditure','Marketing Expenditure']]
x3=sm.add_constant(x2)
result2=sm.OLS(y1,x3).fit()
result2.summary()
```
